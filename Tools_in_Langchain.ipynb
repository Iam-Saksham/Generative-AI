This code cell demonstrates how to use the DuckDuckGoSearchRun tool from LangChain to perform a web search. It first imports the DuckDuckGoSearchRun class, then creates an instance of it named search_tool. Finally, it invokes the search_tool with the query "top news in India today" and prints the results returned by the search.

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_tool = DuckDuckGoSearchRun()

results = search_tool.invoke("top news in India today")
print(results)
```

This code cell prints three pieces of information about the search_tool object, which is an instance of DuckDuckGoSearchRun:

1.search_tool.name: This will output the name of the tool, which is typically a short identifier.

2.search_tool.description: This will output a longer, human-readable description of what the tool does and when it's useful.

3.search_tool.args: This will output a dictionary or schema describing the arguments (inputs) that the tool expects when it is invoked.

```python
print(search_tool.name)
print(search_tool.description)
print(search_tool.args)
```

This code cell demonstrates how to use the ShellTool from LangChain to execute shell commands. It imports ShellTool, initializes an instance named shell_tool, invokes it with the shell command 'ls' (which lists directory contents), and then prints the results of that command.

```python
from langchain_community.tools import ShellTool

shell_tool = ShellTool()

results = shell_tool.invoke('ls')

print(results)
```

This code cell imports the tool decorator from langchain_core.tools. This decorator is essential for converting regular Python functions into LangChain tools, allowing them to be used by LangChain agents and chains.

```python
from langchain_core.tools import tool
```

This code cell redefines the multiply function, adding type hints for its arguments (a: int, b: int) and return value (-> int). This improves readability and helps with validation, which is particularly useful when preparing a function to be used as a LangChain tool.

```python
from langchain_core.tools import tool

# Step 1 - create a function
def multiply(a, b):
    """Multiply two numbers"""
    return a * b
```

This code cell invokes the multiply tool, which was previously defined and decorated with @tool, passing a dictionary {"a": 4, "b": 5} as its arguments. The result of this multiplication (4 * 5 = 20) is then stored in the result variable.



```python
# Step 2 - add type hints
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

This code cell prints the value stored in the result variable, which is the output from the multiply tool invocation from the previous cell. Based on the kernel state, the value of result is currently 20.

```python
# step3: add toll decorator
@tool
def multiply(a: int, b: int) -> int:
   """Multiply two numbers"""
   return a * b
```

```python
result = multiply.invoke({"a": 4, "b": 5})
```

```python
print(result)
```

This code cell prints three key pieces of information about the custom multiply tool:

1.multiply.name: This outputs the name of the tool, which is 'multiply'.

2.multiply.description: This outputs the human-readable description of the tool, taken directly from its docstring: 'Multiply two numbers'.

3.multiply.args: This outputs a dictionary representing the JSON schema of the arguments the multiply tool expects, derived from its type hints. It shows that it expects two integer arguments, 'a' and 'b'.

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```
